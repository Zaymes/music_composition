```python
import glob
midiFile_l = glob.glob("maestro-v1.0.0/**/*.midi")
print(glob.glob("maestro-v1.0.0/**/*.midi"))
```

```python
midiFile_l = midiFile_l[:5]
len(midiFile_l)
```

```python
n_x = 100
max_T_x = 1000
sequence_length = 20
T_y_generated = 200
```

```python
X_list = []

for midiFile in midiFile_l:
  midi_data = pretty_midi.PrettyMIDI(midiFile)
  note_l = [note.pitch for note in midi_data.instruments[0].notes]
  print(max(note_l))
  T_x = len(note_l)
  # print(T_x)
  if T_x > max_T_x:
    T_x = max_T_x
  X_ohe = np.zeros((T_x, n_x))
  # print(X_ohe.shape)
  # print(max(note_l))
  for t in range(T_x): 
    X_ohe[t][note_l[t]-1] = 1
  # add to the list  
  X_list.append(X_ohe)
  # print(X_ohe)
  # print(midi_data.instruments[0].notes)

len(note_l)
print(len(X_list))
print(X_list[0].shape)
print(X_list[1].shape)
print(X_list[2].shape)
```

```python
plt.figure(figsize=(16, 6))
plt.imshow(X_list[1].T, aspect='auto')
plt.set_cmap('gray_r')
plt.grid(True)
```

```python
X_train_list = []
y_train_list = []

sequence_length = 20

for i in range(0, len(X_list)):  
  for j in range(0, len(X_list[i]) - sequence_length):
    X_train_list.append(X_list[i][j:j + sequence_length])
    y_train_list.append(X_list[i][j + sequence_length])

X_train = np.asarray(X_train_list)
y_train = np.asarray(y_train_list)

print("X_train.shape:", X_train.shape)
print("y_train.shape:", y_train.shape)
```

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Input, Dense, LSTM, Flatten, Dropout, Activation
```

```python
# K.clear_session()


model = Sequential()
model.add(LSTM(units = 256, input_shape = (20, 100), return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(units = 256, return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(units = 256))
model.add(Dropout(0.3))
model.add(Dense(256))
model.add(Dropout(0.3))
# model.add(Activation('softmax'))
model.add(Dense(100, activation='softmax'))


print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=20, batch_size=64)
```

```python
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

```python
start = np.random.randint(0, len(X_train_list)-1)
pattern = X_train_list[start]
print(start)
print(pattern.shape)
print(np.expand_dims(pattern, 0).shape)
```

```python
note_l = []

prediction_l = []

new_pattern = np.expand_dims(pattern, 0)

# generate T_y_generated notes
for note_index in range(0, T_y_generated,5):
    pred = model.predict(new_pattern, batch_size=None, verbose=0, steps=None)
    note = np.argmax(pred[0])
    note_pred = np.zeros(100)
    note_pred[note] = 1
  
    new_pattern = np.append(new_pattern[0][1:], [note_pred], axis=0)
    new_pattern = np.expand_dims(new_pattern, 0)
    
    note_l.append(note)
    prediction_l.append(note_pred.tolist())
    
    note_index += note_index
```

```python
print(note_l)
```

```python
new_midi_data = pretty_midi.PrettyMIDI()
cello_program = pretty_midi.instrument_name_to_program('Cello')
cello = pretty_midi.Instrument(program=cello_program)
time = 0
step = 0.3
for note_number in note_l:
    myNote = pretty_midi.Note(velocity=100, pitch=note_number, start=time, end=time+step)
    cello.notes.append(myNote)
    time += step
new_midi_data.instruments.append(cello)
%matplotlib inline

audio_data = new_midi_data.synthesize()
IPython.display.Audio(audio_data, rate=44100)
```
